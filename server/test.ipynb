```python
def merge_data(df_csv, df_json):
    """Merge CSV and JSON data based on matching fields."""
    # Identify common columns for merging
    common_columns = ['gruppe', 'kurzname', 'langtext', 'info', 'lagerort', 'labelIds']
    
    # Create copies to avoid modifying original DataFrames
    df_csv_copy = df_csv.copy()
    df_json_copy = df_json.copy()
    
    # Filter JSON DataFrame to include only common columns
    df_json_filtered = df_json_copy[common_columns]
    
    # Merge DataFrames
    # Using outer merge to keep all records from both datasets
    merged_df = pd.merge(
        df_csv_copy,
        df_json_filtered,
        on=['kurzname'],  # Merge based on kurzname as it appears to be the unique identifier
        how='outer',
        suffixes=('_csv', '_json')
    )
    
    # For each common column (except kurzname which was used for merging),
    # combine the values, preferring the JSON data when available
    for col in common_columns:
        if col != 'kurzname':  # Skip the merge key
            csv_col = f"{col}_csv"
            json_col = f"{col}_json"
            if csv_col in merged_df.columns and json_col in merged_df.columns:
                merged_df[col] = merged_df[json_col].fillna(merged_df[csv_col])
                # Drop the separate columns
                merged_df = merged_df.drop([csv_col, json_col], axis=1)
    
    return merged_df
```
